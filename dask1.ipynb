```python
import csv
import random

# Number of rows
num_rows = 1000000

# Sample data
departments = ["HR", "IT", "Finance", "Marketing", "Sales", "Operations"]
names = ["Ravi", "Sita", "Arjun", "Priya", "Kiran", "Anjali", "Rahul", "Sneha", "Vamsi", "Divya"]

# Create CSV file
with open("employees_1million.csv", mode="w", newline="") as file:
    writer = csv.writer(file)

    # Write header
    writer.writerow(["id", "name", "age", "salary", "department"])

    # Write rows
    for i in range(1, num_rows + 1):
        emp_id = i
        name = random.choice(names)
        age = random.randint(20, 60)
        salary = random.randint(20000, 150000)
        department = random.choice(departments)

        writer.writerow([emp_id, name, age, salary, department])

print("CSV file generated successfully with 1 million rows!")
```
